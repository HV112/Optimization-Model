# Task 4: Optimization Model

## Problem Statement
A factory produces two products: Product A and Product B.

Each product requires machine hours and labor hours.
The factory has limited machine and labor availability.

The goal is to **maximize total profit** while staying within resource limits.


```python
!pip install pulp
```

```python
import pulp
```

```python
model = pulp.LpProblem("Factory_Profit_Maximization", pulp.LpMaximize)
```

```python
x = pulp.LpVariable('Product_A', lowBound=0)
y = pulp.LpVariable('Product_B', lowBound=0)
```

```python
model += 40 * x + 30 * y, "Total_Profit"
```

```python
model += 2 * x + y <= 100, "Machine_Constraint"
model += x + y <= 80, "Labor_Constraint"
```

```python
model.solve()
```

```python
print("Status:", pulp.LpStatus[model.status])
print("Units of Product A:", x.varValue)
print("Units of Product B:", y.varValue)
print("Maximum Profit:", pulp.value(model.objective))
```

## Insights
The optimal number of products is chosen such that
the factory uses available resources efficiently.

The solution maximizes profit while satisfying
machine and labor constraints.
